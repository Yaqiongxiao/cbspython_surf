## loading data

mgz files store data in 4 dimensions (x,y,z,time). In template surface space, each node on the cortex is assigned to an index, so we only need 2 diminesions (node by time). You can use squeeze() to discard empty dimensions.

side note: above, we're using subject data that has been resampled into fsaverage5 template space -- useful, for instance, in group-level analyses. The following examples on anatomy, e.g. labels and surface geometry, are shown for the template surface. They work just the same for subject surfaces in native freesurfer space. Will see later how you can resample subject data to template space.

### surface geometry

If you'd like to look at the geometry of your surface, you can you the freesurfer surf file. It gives you (1) the 3D coordinates of each node on the surface and (2) the faces (nodes that form a triangle) on the triangular surface. 

```python
surf_file = '%s/freesurfer/fsaverage5/surf/lh.orig' % dataDir

surf_coords = nib.freesurfer.io.read_geometry(surf_file)[0]
surf_faces = nib.freesurfer.io.read_geometry(surf_file)[1]

print 'index 0 gives a node by coordinate array: ' + str(surf_coords.shape)
print 'index 1 gives a triangle by node array: ' + str(surf_faces.shape)
```

## saving data

### saving data as an image object

```python
temp_file = '%s/temp.mgz' % yourDir

# load your data, do something to it and save the image.
img1 = nib.load(rest_surf_file)
data1 = img1.get_data()
print data1[0,0,0,0]
data1[0,0,0,0] = .25
img1.to_filename(temp_file)

# check out the saved image
data2 = nib.load(temp_file).get_data()
print data2[0,0,0,0]
```

Since your data was cached (by get_data()), all changes you made are tracked while cached. When you save the image, those changes will be saved too. If, for some reason, you don't want to save your changes, use uncache() before saving.
For more information, see nibabel documentation http://nipy.org/nibabel/images_and_memory.html

```python
# load your data, do something to it and uncache before saving the image.
img1 = nib.load(rest_surf_file)
data1 = img1.get_data()
print data1[0,0,0,0]
data1[0,0,0,0] = .25
img1.uncache()
img1.to_filename(temp_file)

# check out the saved image
data2 = nib.load(temp_file).get_data()
print data2[0,0,0,0]
```

## visualization

Underneath is a function you can use for plotting your data on a brain. 

```python
def viz(data, datamin, datamax, foci=False, labelfile=False, hemi='lh', filename='temp.png'):
    
    brain = Brain('fsaverage5', hemi, 'pial', curv=False, subjects_dir= "%s/freesurfer" % dataDir,
                  config_opts={"background": "white"})
    brain.add_data(data, datamin, datamax, colormap="hot", alpha=0.6)
    
    if labelfile:
        brain.add_label(labelfile, borders=True, color='white')
    if foci:
        brain.add_foci(foci, coords_as_verts=True, scale_factor=.4, color='white')
   
    brain.save_montage(filename, order=['lat', 'med'], orientation='h', border_size=10)
    display(Image(filename=filename, width=700))
```

### projecting data from volume to surface

```python
import nipype.interfaces.freesurfer as fs

vol2surf = fs.SampleToSurface(subjects_dir = '%s/freesurfer' % dataDir,
                              subject_id = 'subject',
                              hemi= 'lh', 
                              source_file = '%s/resting/rest_preprocessed.nii.gz' % dataDir,
                              target_subject = 'fsaverage5',
                              out_file = '%s/rest_surf.mgz' % yourDir,
                              reg_header = True,
                              sampling_method = 'average',
                              sampling_range = (0.2, 0.8, 0.1),
                              sampling_units = 'frac')

res = vol2surf.run() 
```

You can use nipype to run this in a pipeline:
